```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\\Users\Student\Desktop\pic\x.jfif")
hsvimg=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
cv2.imshow("hsv",hsvimg)
cv2.imshow("hsv channel",hsvimg[:,:,0])
cv2.imshow("Saturation channel",hsvimg[:,:,1])
cv2.imshow("Value channel",hsvimg[:,:,2])
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=[20,5])
plt.subplot(1,3,1);plt.imshow(hsvimg[:,:,0],cmap="gray");title=("hsv channel");
plt.subplot(1,3,2);plt.imshow(hsvimg[:,:,1],cmap="gray");title=("Saturation channel");
plt.subplot(1,3,3);plt.imshow(hsvimg[:,:,2],cmap="gray");title=("Value channel");
plt.show()
```

```python
s=cv2.merge([hsvimg[:,:,0],cv2.add(hsvimg[:,:,1],50),hsvimg[:,:,2]])
s=cv2.cvtColor(s,cv2.COLOR_HSV2BGR)
plt.figure(figsize=[12,5])
plt.subplot(1,2,1)
plt.imshow(s[...,::-1])
plt.subplot(1,2,2)
plt.imshow(img[...,::-1])
plt.show()
```
